```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\user\Desktop\機器學習\final project\data\loan.csv")
```

```python
# 計算總筆數的一半
half_count = len(df) / 5

# 計算每個欄位的缺失值數量
missing_counts = df.isnull().sum()

# 找出缺失值數量大於總筆數一半的欄位
columns_with_many_missing = missing_counts[missing_counts > half_count].index

print("欄位缺失值大於總筆數80%的欄位:")
print(columns_with_many_missing)

# 刪除這些欄位
df_dropped = df.drop(columns=columns_with_many_missing)
```

```python
def contains_non_numeric(series):
    try:
        series_as_numeric = pd.to_numeric(series)
        return False
    except ValueError:
        return series.astype(str).str.contains('[a-zA-Z]').any()

# 找出非數字值的欄位
text_encoded_columns = [col for col in df_dropped.columns if contains_non_numeric(df_dropped[col])]

print("包含非数字值的欄位:")
for column in text_encoded_columns:
    print(column)
```
